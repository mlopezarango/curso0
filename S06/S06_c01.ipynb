```python

## Funciones de Numpy para los archivos I/O
En muchos casos leer y guardar datos para procesar y trabajar en numpy es complicado usando funciones nativas de python como *open*, para procesar datos en input/output tenemos algunas funciones.
 
La función para leer texto es:

```python
loadtxt(fname, dtype=numpy.float,
 comments="#", delimiter=None, 
 converts=None, skiprows=0, 
 usecols=None, unpack=False, ndim=0)
```

La función para guardar texto es:

```python
savetxt(fname, X, fmt='%.18e', 
delimiter='', newline='\n',
header='', footer='', comments='#')
```


````
import numpy as np

#Valores dependientes

c1 = np.linspace(0,1,10)
c2 = c1**2
c3 = 2*c1
c4 = np.sin(c2)
c5 = np.cos(c3)

table = np.array([c1,c2,c3,c4,c5])

np.savetxt('datos.dat',table.T,fmt='%.6e',header='c1 \t c2 \t c3 \t c4 \t c5',footer='By: Sheryl Avendaño')

with open("datos.dat") as filedat:
  print(filedat.read())

p1,p2,p3,p4,p5 = np.loadtxt('datos.dat',unpack=True, skiprows=5)

#print('¿p1 = c1?', p1[1]==c1[1],p1[1],c1[1])
print(p1)
````

#**Operaciones Vectorizadas: Optimizaciones computacionales en los arrays de `Numpy`:**


Los arreglos multidimendsionales de `NumPy` pueden utilizarse en expresiones matemáticas para realizar cálculo utilizando las entradas de una matriz. En general, se implementan funciones matemáticas que operan sobre cada una de las entradas de la matriz. 

A continuación, realizaremos varias operaciones, a través de ejercicios corto, como:

1. Elevar al cuadrado cada elemento de una matriz dada.
2. Tomar la raiz cuadrada de cada uno de los elementos de una matriz.
3. Sumar a cada elemento de una matriz un valor dado.
4. Sumar elemento a elemento dos matrices.
5. Sumar todos los elementos de una matriz. 
6. Sumar a lo largo de una dirección los elementos de una matriz.

Cada una de estas operaciones representa una operación vectorizada.

Recordemos que los array de `NumPy` son homogéneos: **un array solo puede contener datos de un solo tipo**. Por ejemplo, una matriz puede contener enteros de 8 bits o números de coma flotante de 32 bits, pero no una mezcla de los dos. Por el contrario, las listas y tuplas de `Python` no tienen restricciones en la variedad de tipos que pueden contener; una lista dada podría contener simultáneamente cadenas, enteros y otros objetos. Esta restricción en el contenido de tipos en un array tiene un gran beneficio; al "saberse" que los contenidos de un array son homogéneos en el tipo de datos, `NumPy` puede delegar la tarea de realizar operaciones matemáticas en los contenidos de dicho arreglo a un código C compilado y optimizado. Este es un proceso que se conoce como vectorización. El resultado de esto puede ser una tremenda aceleración en relación con el cálculo análogo realizado en Python, que debe verificar minuciosamente el tipo de datos de cada uno de los elementos a medida que itera sobre las matrices, ya que Python generalmente trabaja con listas con contenido sin restricciones de en los tipos de los datos que alberga.

Defina una función en `Python` que sume los elementos de un array a lo largo de las filas y nombrela `sum_along_rows`.

````
# Escriba la solución aquí.
````

````
#@title Solución oculta al ejercicio
def sum_along_rows(a = np.array([[1,2],[3,4]])):
  row_sum = np.zeros(a.shape[1], dtype=float)
  for i in range(a.shape[0]):
    for j in range(a.shape[1]):
      row_sum[i] += a[i,j]
  return row_sum

print('Matriz y\n', y, '\n')
print('sum_along_rows(y)', sum_along_rows(y))
````

Use `set` para devolver los valores únicos para la cantidad de cilindros que tienen los autos en nuestro conjunto de datos.

````
cylinders = set(d['cyl'] for d in mpg)
cylinders
````

Aquí hay un ejemplo más complejo en el que estamos agrupando los autos por número de cilindros y encontrando el promedio de mpg en ciudad para cada grupo.

````
CtyMpgByCyl = []

for c in cylinders: # iterate over all the cylinder levels
    summpg = 0
    cyltypecount = 0
    for d in mpg: # iterate over all dictionaries
        if d['cyl'] == c: # if the cylinder level type matches,
            summpg += float(d['cty']) # add the cty mpg
            cyltypecount += 1 # increment the count
    CtyMpgByCyl.append((c, summpg / cyltypecount)) # append the tuple ('cylinder', 'avg mpg')

CtyMpgByCyl
````

##Series
Las series de pandas son una de las estructuras centrales en para Pandas. Puede pensarlas como un cruce entre una lista y un diccionario. Los items son guardados en orden y hay etiquetas con las cuales se puede recuperar los items.

![alt text](https://github.com/tomasate/Curso0/blob/master/Captura2.PNG?raw=true)

Una forma de visualizarlo es como 2 columnas de datos, el primero es un índice especial, tal y como el `key` de un diccionario y el segundo son los datos.
Notemos que la columna de los datos tiene una etiqueta por sí misma, y puede consultarse usando el atributo `.name` esto es diferente que con los diccionarios, y es bastante útil a la hora de mezclar diferentes tablas de datos.

Por ahora intentemos crear una serie nosotros.

````
animals = ['Tiger', 'Bear', 'Moose']
pd.Series(animals)
````

````
numbers = [1, 2, 3]
pd.Series(numbers)
````

Pandas usa por debajo el módulo numpy por lo que permite usar objetos como `None` en caso de querer evitar llenar un espacio en el caso de strings. En el caso de números automáticamente crea una variable de tipo float llamada `NaN`

````
animals = ['Tiger', 'Bear', None]
pd.Series(animals)
````

````
numbers = [1, 2, None]
pd.Series(numbers)
````

También es posible hacer una serie a partir de diccionarios en lugar de listas, en este caso el `key` se convierte en nuestro índice

````
sports = {'Archery': 'Bhutan',
          'Golf': 'Scotland',
          'Sumo': 'Japan',
          'Taekwondo': 'South Korea'}
s = pd.Series(sports)
s
````

````
s.index
````

También podemos hacer la lista de los datos y pasar el índice por separado como otra lista.

````
s = pd.Series(['Tiger', 'Bear', 'Moose'], index=['India', 'America', 'Canada'])
s
````

###Consultar una serie
Es posible consultar en una serie bien sea por la posición del índice o bien por la etiqueta del índice. Para consultar por valor numérico, empezando por cero utilizamos el atributo `.iloc[]` para consultar por la etiqueta del índice usamos el atributo `.loc[]`


````
sports = {'Archery': 'Bhutan',
          'Golf': 'Scotland',
          'Sumo': 'Japan',
          'Taekwondo': 'South Korea'}
s = pd.Series(sports)
s
````

````
s.iloc[3]
````

````
s.loc['Taekwondo']
````

Pandas está diseñado de una forma tal que sea amigable buscar este tipo de objetos por el operador indexación, por lo que si buscamos `s[i]` con `i` siendo un número, él interpreta de forma subyacente lo que sería el operador `iloc`.

````
s[3]
````

````
s['Golf']
````

Sin embargo hay que tener cuidado cuando por ejemplo se usen números como índices, pues no necesariamente el índice debe coincidir con la etiqueta, por lo que siempre es mejor usar los atributos.

````
sports = {99: 'Bhutan',
          100: 'Scotland',
          101: 'Japan',
          102: 'South Korea'}
s = pd.Series(sports)
s[0]
````

````
s.iloc[0]
````

Consideremos una tarea simpleque implique tener en cuenta todos los valores dentro de una serie, y operar con ellos, esto puede ser encontrar un cierto número, resumir datos, o transformarlos de alguna forma. Una aproximación típica a esto es la iteración sobre todos los items de la serie e invocar la operación que nos interesa, por ejemplo, creemos una serie de floats que bien podrían ser precios para ciertos productos.

````
s = pd.Series([100.00, 120.00, 101.00, 3.00])
s
````

````
total = 0
for item in s:
    total+=item
print(total)
````

O bien podríamos usar lo que sabemos sobre vectorización

````
import numpy as np

total = np.sum(s)
print(total)
````

Comprobemos cuál de los 2 métodos es mejor:

````
#Creamos una serie de 10000 números enteros entre 0 y 1000 
s = pd.Series(np.random.randint(0,1000,10000))
s.head()
````

Usemos la funcion `timeit` que correrá nuestro código un par de veces y nos dirá qué tanto tiempo tarda en ejecutarlo

````
%%timeit -n 100
summary = 0
for item in s:
    summary+=item
````

````
%%timeit -n 100
summary = np.sum(s)
````

Aplica algo similar para el caso en el que deseamos modificar los datos de la serie

````
%%timeit -n 10
s = pd.Series(np.random.randint(0,1000,10000))
for label, value in s.iteritems():
    s.loc[label]= value+2
````

````
%%timeit -n 10
s = pd.Series(np.random.randint(0,1000,10000))
s+=2
````

Podemos agregar un nuevo item a nuestra serie de la siguiente forma

````
s = pd.Series([1, 2, 3])
s.loc['Animal'] = 'Bears'
s
````

Hasta ahora hemos visto ejemplos en los que el índice de los datos es único, veamos un ejemplo en el cual los índices NO son únicos, lo cual hace de las seriesun objeto fundamentalmente diferente a las bases de datos que suelen encontrarse frecuentemente.

````
original_sports
````

Note que la serie original_sports no fue modificada.

##Data Frame o Cuadro de Datos
El data frame es una estructura de datos que podría bien ser considerada el corazón de el módulo de Pandas. Es el objeto primario con el cual un analista de datos en python puede encontrarse trabajando.
Uno puede pensar en un data frame como una serie dos dimensional, donde tenemos un índice y múltiples columnas, donde cada columna tendrá una etiqueta. De hecho la distinción entre una columna. De hecho uno puede pesar que la diferencia entre una columna y una fila es solo conceptual, lo que haría de un Data frame una suerte de arreglo 2 dimensional.

Existen multiples formas de hacer un data frame, podemos usar un grupo de series, donde cada representaría una fila de datos. También podríamos usar diccionarios

````
purchase_1 = pd.Series({'Nombre': 'Tomas',
                        'Objeto Comprado': 'Comida para perro',
                        'Precio': 22.50})
purchase_2 = pd.Series({'Nombre': 'Sheryl',
                        'Objeto Comprado': 'Arena de gatos',
                        'Precio': 12.50})
purchase_3 = pd.Series({'Nombre': 'Sebastian',
                        'Objeto Comprado': 'Semillas para ave',
                        'Precio': 5.00})
df = pd.DataFrame([purchase_1, purchase_2, purchase_3], index=['Tienda 1', 'Tienda 1', 'Tienda 2'])
df.head()
````

Similar a las series podemos extraer información usando los tributos `loc` y `iloc`sin enmbargo en este caso como el data frame es 2 dimensional nos entregará información 2 dimensional (en ocasiones 1 dimensional), es decir un data frame (o una serie)

````
df.loc['Tienda 1']
````

````
df.loc['Tienda 2']
````

Algo es el hacer una transposición del data frame, es decir cambiar las filas por las columnas

````
df.T
````

Ahora supongamos que estamos interesados en ver los precios en nuestro data frame

````
df['Precio']
````

Inclusive si quisieramos ver los costos de la tienda número 1 tendríamos.

````
df.loc['Tienda 1']['Precio']
````

Esto último sería encadenar operadores de indexación y es muy útil a la hora de hacer consultas. Es importante tener en cuenta que esto termina por mostrar una copia del data frame, diferente a una vista del data frame. Esto es algo a considerar si estoy modificando mis datos.

Podemos presentar además solo las columnas que deseamos de la siguiente forma

````
df.loc[:,['Nombre', 'Precio']]
````

Veamos cómo podemos eliminar información de un data frame

````
df.drop('Tienda 1')
````

Lo anterior no ha cambiado nuestro data frame

````
df
````

Lo mejor es hacer una copia con la modificación

````
copy_df = df.copy()
copy_df = copy_df.drop('Tienda 1')
copy_df 
````

Finalmente podríamos agregaruna nueva columna

````
df['Location'] = None
df
````
