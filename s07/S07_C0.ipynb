````
#@title Esto es Python para programación web
import pandas as pd
from sodapy import Socrata
from datetime import date

def conectapp(dominio,dataset_id,token=None):
  with Socrata(dominio,token) as client:
    try:
      metadata = client.get_metadata(dataset_id)
    except Exception as e:
      print("Hay un error al iniciar el cliente o acceder a la base de datos.")
      print("Esta es la respuesta:",e)
    else:
      createdAt = date.fromtimestamp(metadata['createdAt'])
      updateAt = date.fromtimestamp(metadata['viewLastModified'])
      print("La base de datos fue creada en:",createdAt,",su ultima modificación está registrada en",updateAt)
    finally:
      return client,metadata

def sodatoDataFrame(dominio,dataset_id,token=None,limit = 5000):
  client,metadata = conectapp(dominio,dataset_id,token=token)
  nameFirstCol = metadata['columns'][0]['fieldName']
  rows_id = int(client.get(dataset_id, select="COUNT({0})".format(nameFirstCol))[0]['COUNT_{0}'.format(nameFirstCol)])
  print("El número de datos identificados es:", rows_id)
  isMoreData = rows_id > limit 
  results = client.get(dataset_id,limit=limit)
  if not isMoreData:
    dataset = pd.DataFrame.from_records(results)
  else:
    delta,start,i = limit,limit,0  
    dataset_list = [pd.DataFrame.from_records(results)]
    row_current = limit
    while isMoreData:
      start += delta*i
      results = client.get(dataset_id,limit=limit,offset=start)
      dataset_list.append( pd.DataFrame.from_records(results) )
      i +=1
      row_current += limit*i
      isMoreData = rows_id > row_current
    dataset = pd.concat(dataset_list,ignore_index=True)
  return dataset
````

#**Oparaciónes de agregación y agrupamiento**

## **Introducción**

La categorización de un conjunto de datos y la aplicación de una función a cada grupo, ya sea una agregación o transformación, a menudo es un componente crítico de un flujo de trabajo de análisis de datos. Después de cargar y preparar un dataset, una tarea común es calcular estadísticas de grupo o tablas dinámicas para fines de informes o visualización. `pandas` proporciona una herramienta de agrupado flexible y de alto rendimiento, que le permite cortar, cortar en grupos y resumir conjuntos de datos de una manera natural.

Como veremos a continuación, con la funcionalidad y el poder de `Python` y `pandas`, podemos realizar agrupamientos complejos, utilizando cualquier cualquier función que acepte un objeto pandas o un arreglo `NumPy`. En este capítulo, aprenderá a:

##**Método `GroupBy`**

En lo subsecuente, trabajaremos con el concepto de **dividir-aplicar-combinar** para hablar sobre operaciones grupales. En la primera etapa del proceso, los datos contenidos en un objeto pandas, ya sea una Serie, un `DataFrame` o de otro modo, se dividen en grupos en función basados en una o más claves que se proporcionen. La división se realiza en un eje particular de un objeto. Por ejemplo, un `DataFrame` se puede agrupar en sus filas (`axis=0`) o sus columnas (`axis= 1`). Una vez hecho esto, se aplica una función a cada grupo, produciendo un nuevo valor. Finalmente, los resultados de todas esas aplicaciones de funciones se combinan en un objeto final. La forma del objeto resultante generalmente dependerá de lo que se haga con los datos. 

![texto alternativo](https://github.com/diplomado-bigdata-machinelearning-udea/curso0/blob/master/s07/imagenes/group_process.png?raw=true)

Cada clave de agrupamiento puede tomar muchas formas, y las claves no tienen que ser todas del mismo tipo:

- Una lista o arreglo de valores que es de la misma longitud que eje que se está agrupando.

- Un valor que indica el nombre de una columna en un `DataFrame`

- Un `dict` o una `Series` que proporciona una correspondencia entre los valores del eje y los nombres de los grupos.

- Una función que se invocará en el índice del eje o en las etiquetas individuales del índice.

Nótese que los últimos tres métodos son simplemente "atajos" para producir un arreglo de valores que será usado para dividir el objeto. A lo largo de la presenta sección iremos brindando ejemplos de estos métodos. 

````
df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],'key2' : ['uno', 'dos', 'uno', 'dos', 'uno'],
                'data1' : np.random.randn(5), 'data2' : np.random.randn(5)})

df
````

Sumpongamos que se quiere calcular el valores medio de la columna `data1` usando los grupos de etiquetas `key1`. Existen muchas formas de hacer esto. Una de ellas, consiste en acceder a `data1` y aplicar el método `groupby` con la columna (de tipo `Series`) en la clave `key1`:

````
grouped = df['data1'].groupby(df['key1'])

grouped
````

La variable `grouped` se conoce como un objeto `GroupBy`. No se ha calculado nada aun, a excepción de de unos datos intemedios sobre laa clave de grupo `df['key1']`. La idea es que este objeto tenga toa la información necesaria para luego aplicar alguna operación a cada uno de los grupos. Por ejemplo, para calcular la media por grupo, podemos llamar el método `mean()` del objeto `GroupBy`:

````
grouped.mean()
````

Nóteses que, lo importante aquí es que los datos (un objeto `Series`) han sido agrupados de acuerdo con los valores de la clave 'key1'. Se hace evidente, además, que el índice del objeto `Series` resultante tiene el nombre `key1` ya que se realizo mediante la especificación de `df['key1']`.


Veamos que, efectivamente, el tipo de dato devuelve el método que acabamos de aplicar es `Series`:

````
type(grouped.mean())
````

Si pasamos una lista de varios arreglos, el resultado es diferente:

````
means = df['data1'].groupby([df['key1'], df['key2']]).mean()
means
````

En este caso, hemos agrupado los datos usando dos claves. Se evidencia que objeto `Series` posee una jerarquización en el indexado de las parejas únicas de las claves observadas.

Usemos el método `unstack()`, para hacer una combinación de todas las claves y visualizar los valores medios en cada caso:

````
means.unstack()
````

En estos ejemplos, las claves son todas `Series`; téngase en cuenta que pueden ser cualquier array de la misma longitud de la tabla aprpropiada 

````
len(df)
````

Si usamos otra serie, o arreglo, con una longitud de cinco elementos, podremos agrupar nuestro `DataFrame` `df`:

````
departamentos = np.array(['Antioquia', 'Choco', 'Cundinamarca', 'Amazonas', 'Vaupes'])
tiempo = np.array([2008, 2008, 2009, 2008, 2009])

df['data1'].groupby([departamentos, tiempo]).mean()
````

Con frecuencia, la información de agrupación se encuentra en el mismo DataFrame que los datos en los que desea trabajar. En ese caso, puede pasar los nombres de columna (ya sean cadenas, números u otros objetos de Python) como las claves del grupo:

````
df.groupby('key1').mean()
````

````
 df.groupby(['key1', 'key2']).mean()
````

Nótese que en el primer caso `def.groupby('key1').mean()` no contiene columna `key2` en su resultado. Lo anterior se debe a que `df['key2']` no es de tipo numérico, decimos que es una columna 'problemática', y por lo tanto se excluye del resultado. De forma predeterminada, todas las columnas numéricas son agregadas, aunque es posiblre filtrarlas a un subconjunto menor, como se verá más adelante.

Independientemente del objetivo al usar `groupby`, un método `GroupBy` generalmente útil corresponde al tamaño que retorna una serie que contiene tamaños de los grupo:

````
df.groupby(['key1', 'key2']).size()
````

##**Iteración sobre Grupos:**

El objeto `GroupBy` admite la iteración, generando una secuencia de 2 tuplas que contienen el nombre del grupo junto con la porción de datos. Considere el siguiente conjunto de datos de ejemplo:

````
for nombre, grupo in df.groupby('key1'):
  print("Nombre:", nombre)
  print(grupo)
````

En el caso de claves múltiples, el primer elemento de la tupla será una tupla de los valores de las claves:

````
for (k1, k2), group in df.groupby(['key1', 'key2']):
  print("Claves: ",k1,',', k2)
  print(group)
````

En general, se puede realizar cualquier tipo de operación con los datos. Una aplicación que puede resultar útil, es el cómputo de un `dict` de las partes de un conjunto de datos como una línea

````
dict(list(df.groupby('key1')))
````

````
partes = dict(list(df.groupby('key1')))
````

````
partes['a']
````

````
partes['b']
````

Por defecto, `groupby` agrupa a lo largo de `axis=0`, per puede agruparse en cualquier eje. Ppor jemplo, podríamos agrupas las columnas de nuestro ejemplo de acuerdo a los tipos de datos `dtypes` que almacena:

````
df.dtypes
````

````
grouped = df.groupby(df.dtypes, axis=1)
dict(list(grouped))
````

## **Selección de una Columna o subconjunto de columnas**

El indexado de un objeto `GroupBy`, el cual es creado a partir de un `DataFrame` con nombres de columnas o arreglos de nombres de columnas tiene el efecto de seleccionar esas columnas para la agregación. Lo anterior sigifica que



```
df.groupby('key1')['data1']
df.groupby('key1')[['data2']]
```
Es una sintaxis 'más' intuitiva que:

```
df[['data1']].groupby(df['key1'])
df[['data2']].groupby(df['key1'])
```

la cual es aceptable también.

En específico, para grandes conjuntos de datos, puede resultar conveniente agregar solo unas pocas columnas. Por ejemplo, en el conjunto de datos anterior, para calcular los valores medios solo de las columnas `data2` y obtener el resultado como un `DataFrame`, escribimos:


````
df.groupby(['key1', 'key2'])[['data2']].mean()
````

El objeto que se devuelve en esta operación de indexación es un `DataFrame` agrupado siempre que se pase una lista o arreglo junto a un método como `mean` y una Serie agrupada si solo se pasa un nombre de la columna o lista de ellas:

````
 series_grouped = df.groupby(['key1', 'key2'])['data2']
 series_grouped
````

````
series_grouped.mean()
````

##**Agrupamiento con funciones**

Cualquier función que se pase como clave se llamará una vez por valor de índice, y los valores devueltos se utilizarán como nombres de grupo. En concreto, consideremos el `DataFrame` anterior, cuyo índice corresponde a nombres de personas. Supongamos que queremos agrupar por longitud de los nómbres; se ppodría calcular un areglo de longitudes de los nombres, pero en su lugar, le podemos pasar al método `groupby` la función `len`:

````
personas.groupby(len).sum()
````

Si realizamos una mezcla de funciones con arreglos, diccionarios o series, todo es convertido internamente en un arreglo:

````
lista_claves = ['uno', 'uno', 'uno', 'dos', 'dos']
personas.groupby([len,lista_claves]).min()
````

##**Agrupamiento por niveles de índice o indexado múltiple**

De forma conveniente, podemos usar esta jerarquización para agrupar los datos usando uno de los niveles de un eje de un índice. Para realizar esto, se pasa el número de nivel o el nombre, usando la palabra clave `level`. Vemos esto con un ejemplo.

````
columnas = pd.MultiIndex.from_arrays([['COL', 'COL', 'COL', 'VEN', 'VEN'],[1, 3, 5, 1, 3]], names=['pais', 'mes'])
````

df1 = df.copy()
df1['Recuperacion'] = df['Recuperado']-df['Contagiado']
s = df1.groupby('Recuperacion')['Edad'].count()
